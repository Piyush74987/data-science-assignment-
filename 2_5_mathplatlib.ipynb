## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
Ans-> Matplotlib is a popular data visualization library in Python. It provides a wide range of tools for creating various types of static, interactive, and animated visualizations. Matplotlib is used to visualize data and gain insights from it, making it easier to understand complex trends, patterns, and relationships in your data.

Key features and reasons for using Matplotlib:

Easy-to-Use Interface: Matplotlib's simple and intuitive interface allows you to create high-quality plots with just a few lines of code.

Wide Variety of Plot Types: Matplotlib offers a variety of plot types to suit different data visualization needs, ranging from simple line plots to more complex 3D visualizations.

Customization: You can customize almost every aspect of your plots, including colors, labels, titles, axes, legends, and more.

Publication-Quality Plots: Matplotlib produces publication-quality plots suitable for both digital and print media.

Integration: Matplotlib can be easily integrated with other libraries like NumPy, Pandas, and SciPy, making it a powerful tool for data analysis and visualization.

Support for Different Output Formats: Matplotlib supports various output formats, such as PNG, PDF, SVG, and more.

Five common plots that can be created using the Pyplot module of Matplotlib:

Line Plot: Display data points connected by straight line segments, useful for visualizing trends over continuous data.

Bar Plot: Represent data using rectangular bars, often used to compare quantities of different categories.

Scatter Plot: Display individual data points as dots, suitable for visualizing the relationship between two variables.

Histogram: Display the distribution of a dataset by dividing it into intervals and plotting the frequency or count in each interval.

Pie Chart: Display the proportion of different categories in a dataset as slices of a pie, representing relative sizes.

These are just a few examples of the types of plots you can create with Matplotlib's Pyplot module. The library offers many more plot types and customization options to cater to various visualization needs.

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to create multiple plots in a single figure, making it easy to compare and visualize different sets of data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([300, 250, 200, 150, 100, 50])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout and display the plots
# plt.tight_layout()
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot (also known as a box-and-whisker plot) is a graphical representation that displays the distribution of a dataset's summary statistics, including the median, quartiles, and potential outliers. It's particularly useful for visualizing the spread and skewness of data and identifying potential anomalies or outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Data to be plotted
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 5))  # Set the figure size
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot Example')
plt.show()
```
